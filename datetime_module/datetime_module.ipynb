```python
# 1. Get the current date and time, and display it in the format DD/MM/YYYY HH:MM:SS
current_time = datetime.now()
print(f"Current Date and Time: {current_time.strftime('%d/%m/%Y %H:%M:%S')}")
```

```python
#2. Get yesterday's date and print it
yesterday = current_time - timedelta(days=1)
print(f"Yesterday's Date: {yesterday.strftime('%d/%m/%Y')}")
```

```python
# 3. Function to return the number of days remaining until New Year's Day of the next year
def days_until_new_year():
    next_year = current_time.year + 1
    new_year_date = datetime(next_year, 1, 1)
    remaining_days = new_year_date - current_time
    return remaining_days.days

print(f"Days remaining until New Year's Day: {days_until_new_year()}")
```

```python
# 4. Ask the user to enter their date of birth and calculate their exact age in years, months, and days
def calculate_age(birth_date_str):
    birth_date = datetime.strptime(birth_date_str, "%d/%m/%Y")
    today = datetime.now()
    age_years = today.year - birth_date.year
    age_months = today.month - birth_date.month
    age_days = today.day - birth_date.day
    
    if age_months < 0:
        age_years -= 1
        age_months += 12
    if age_days < 0:
        age_months -= 1
        age_days += (birth_date.replace(year=today.year, month=today.month) - timedelta(days=1)).day
    
    return age_years, age_months, age_days

# Example: Ask user for input
dob_input = input("Enter your date of birth (DD/MM/YYYY): ")
age_years, age_months, age_days = calculate_age(dob_input)
print(f"Age: {age_years} years, {age_months} months, {age_days} days")
```

```python
# 6. Given two date strings, "2025-01-10" and "2024-12-20", calculate the number of days, hours, and minutes between them
date_str1 = "2025-01-10"
date_str2 = "2024-12-20"
date1 = datetime.strptime(date_str1, "%Y-%m-%d")
date2 = datetime.strptime(date_str2, "%Y-%m-%d")
time_difference = date1 - date2

days = time_difference.days
hours = time_difference.seconds // 3600
minutes = (time_difference.seconds // 60) % 60
print(f"Time Difference: {days} days, {hours} hours, {minutes} minutes")
```

```python
#Add 100 days to user input date
def add_100_days(date_str):
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")
    return (date_obj + timedelta(days=100)).date()

# Example usage
event_date = "2025-01-01"  # Replace with input()
new_date = add_100_days(event_date)
print("New date after 100 days:", new_date.strftime("%d/%m/%Y"))
```

```python
#Countdown timer until a deadline
def countdown_to(deadline_str):
    deadline = datetime.strptime(deadline_str, "%Y-%m-%d")
    today = datetime.now()
    return (deadline - today).days

print("Days until final project deadline:", countdown_to("2025-12-31"))
```

```python
#Number of weekdays between two dates
def count_weekdays(start_str, end_str):
    start = datetime.strptime(start_str, "%Y-%m-%d")
    end = datetime.strptime(end_str, "%Y-%m-%d")
    delta = timedelta(days=1)
    count = 0
    while start <= end:
        if start.weekday() < 5:
            count += 1
        start += delta
    return count

print("Weekdays between:", count_weekdays("2024-12-01", "2024-12-31"))

```

```python
#Schedule task every 15 days (5 times)
def schedule_tasks(start_str):
    start = datetime.strptime(start_str, "%Y-%m-%d")
    return [start + timedelta(days=15*i) for i in range(5)]

dates = schedule_tasks("2025-01-01")
for d in dates:
    print(d.strftime("%d/%m/%Y"))
```

```python
#Complete months between two dates
from datetime import datetime
from dateutil.relativedelta import relativedelta

def complete_months_between(start_str, end_str):
    start = datetime.strptime(start_str, "%Y-%m-%d")
    end = datetime.strptime(end_str, "%Y-%m-%d")
    delta = relativedelta(end, start)
    return delta.years * 12 + delta.months

print("Complete months between:", complete_months_between("2024-01-01", "2025-07-01"))
```

```python
#Sort ISO timestamps
timestamps = [
    "2025-03-01T14:00:00",
    "2024-11-20T09:30:00",
    "2025-01-10T10:15:00"
]

sorted_timestamps = sorted(timestamps, key=lambda x: datetime.fromisoformat(x))
print("Sorted Timestamps:")
for ts in sorted_timestamps:
    print(ts)
```
